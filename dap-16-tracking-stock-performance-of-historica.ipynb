```python
import pandas as pd
```

```python
#import datasets for stocks being tracked 
apple = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/AAPL.csv')
facebook = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/FB.csv')
google = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/GOOGL.csv')
nvidia = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/NVDA.csv')
tesla = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/TSLA.csv')
twitter = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/TWTR.csv')
```

```python
dfs = [apple, facebook, google,nvidia,tesla,twitter]
```

```python
# add two columns for the moving average 
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA50'] = df.Close.rolling(50).mean()
```

```python
apple.head(50)
```

```python
apple.head(200)
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] -  df['Previous day close price']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in volume'] = df['Volume'] -  df['Previous day volume']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
