```python
import pandas as pd
```

```python
#import datasets for stocks being tracked 
apple = pd.read_csv('/kaggle/input/stock-market-for-nasdaq/Apple.csv')
facebook = pd.read_csv('/kaggle/input/stock-market-for-nasdaq/Facebook.csv')
google = pd.read_csv('/kaggle/input/stock-market-for-nasdaq/Google.csv')
nvidia = pd.read_csv('/kaggle/input/stock-market-for-nasdaq/Nvidia.csv')
tesla = pd.read_csv('/kaggle/input/stock-market-for-nasdaq/Tesla.csv')
twitter = pd.read_csv('/kaggle/input/stock-market-for-nasdaq/Twitter.csv')
```

```python
dfs = [apple, facebook, google,nvidia,tesla,twitter]
```

```python
# add two columns for the moving average 
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA50'] = df.Close.rolling(50).mean()
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] -  df['Previous day close price']
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```
